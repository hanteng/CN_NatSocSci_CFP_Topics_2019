```python
with open ("Topics_List.txt", encoding="utf-8") as fin:
    fin_content = fin.readlines()
```

```python
# Trying list comprenhensions
# [x for x in fin_content if x[0] not in "*0123456789"]
# [x.strip() for x in fin_content if x[0] not in "*0123456789"]
# [y for y in [x.strip() for x in enumerate(fin_content) if x[0] not in "*0123456789"] if len(y)>0]
line_markers = [(i,y) for i,y in [(i, x.strip()) for i,x in enumerate(fin_content) if x[0] not in "*0123456789"] if len(y)>0]
line_markers
```

## Modularized Function

```python
import pandas as pd

def process_data(tryone):
    begin = line_markers[tryone]
    #print (begin)
    try:
        end = line_markers[tryone+1]
        #print (end)
    except:
        end = (len(fin_content)-1,"")
        #print (end)
    data = fin_content[begin[0]+1:end[0]-1]
    # [[y[0].split(" ")[1], y[1].strip(), y[0].split(" ")[0]] for y in [x.split(".") for x in data] if "*" in y[0]]
    data_cleaned = [[y[0].split(" ")[1], y[1].strip(), y[0].split(" ")[0]] if "*" in y[0] else [y[0], y[1].strip(), ""] for y in [x.split(".") for x in data] ]
    # print (data_cleaned)
    df = pd.DataFrame(data_cleaned, columns=['号','课题','具体_方向'])
    df ['学科'] = begin[1]
    return(df)

df_all = pd.DataFrame()
for i in range(23): 
    df = process_data(i)
    df_all = pd.concat([df_all, df])


# dict for 学科
学科_顺序 = df_all[['学科']].drop_duplicates().学科.values
disciplines = {v:i for i,v in enumerate(学科_顺序)} 
 

df_all['学科_顺序'] = [disciplines[d] for d in df_all ['学科']]


df_all['具体'] = [True if "*" in y else False for y in df_all['具体_方向']]

df_all = df_all [['学科_顺序','学科','具体','号','课题','具体_方向']].reset_index()

```

```python
df_all.index.name = "index_all"
```

### 輸出 

```python
df_all.reset_index().to_csv("Topics_List.tsv", sep='\t', index=False, columns=['学科_顺序','学科','具体','号','课题','具体_方向','index','index_all'])
```

## 計量
